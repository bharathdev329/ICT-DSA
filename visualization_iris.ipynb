```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
iris=sns.load_dataset('iris')

df=pd.DataFrame(iris)
df.head()
```

```python
sns.histplot(iris['sepal_length'],bins=10,kde=True)
plt.show()
```

```python
sns.pairplot(iris,hue='species')
plt.show()
```

```python
#boxplot

sns.boxplot(x='species',y='sepal_length',data=iris)
plt.show()
```

A box plot provides a summary of the data through its quartiles and highlights the variability and skewness.
It provides a Five-Number Summary:

Minimum: The smallest data point, excluding outliers.

First Quartile (Q1): The median of the lower half of the dataset (25th percentile).

Median (Q2): The middle value of the dataset (50th percentile).

Third Quartile (Q3): The median of the upper half of the dataset (75th percentile).

Maximum: The largest data point, excluding outliers.

1. Setosa: The sepal lengths are tightly clustered with a small IQR, indicating low variability. There are no significant outliers.
2. Versicolor: The sepal lengths have a larger IQR, indicating more variability. There may be a few outliers.
3. Virginica: This species shows the most variability in sepal lengths with a wide IQR and several outliers.

```python
#violinplot

sns.violinplot(x='species',y='sepal_length',data=iris)
plt.show()
```

A violin plot of sepal_length across different species can show you how the sepal length varies among the species and whether there are significant differences in their distributions.


The white dot in the center of each violin represents the median sepal length
for that species.

The thick black bar in the center shows the interquartile range (IQR), which is the middle 50% of the data.

The thin black line (whiskers) extending from the bar shows the range of the data, excluding outliers.

Outliers, if any, are plotted as individual points outside the whiskers.



1.  Setosa: The distribution is relatively narrow, indicating less variation in sepal length. The median is lower compared to the other species.
2.   Versicolor: The distribution is wider, showing more variation in sepal length. The median is higher than Setosa but lower than Virginica.
3. Virginica: This species has the widest distribution, indicating the most variation in sepal length. The median is the highest among the three species.**bold text**



```python
#swarmplot

sns.swarmplot(x='species',y='sepal_length',data=iris)
plt.show()
```

A swarm plot of the Iris dataset can provide detailed insights into the distribution of sepal lengths across different species

Each dot represents an individual iris flowers sepal length.

The spread of dots shows the range and density of sepal lengths for each species.



1.   Setosa: The sepal lengths are relatively consistent, with most points clustered closely together, indicating less variation.
2. Versicolor: There is more variation in sepal lengths compared to Setosa, with points spread out over a wider range
3. Virginica: This species shows the most variation in sepal lengths, with points spread out even more widely.


Swarm plots are often used alongside box plots or violin plots to provide a more detailed view of the data distribution. While box plots and violin plots summarize the data, swarm plots show all individual data points, giving a fuller picture.


```python
#Stripplot
sns.stripplot(x='species',y='sepal_length',data=iris)
plt.show()
```

1. Setosa: The sepal lengths are tightly clustered, indicating low variability.
2. Versicolor: There is more variation in sepal lengths, with points spread out over a wider range.
3. Virginica: This species shows the most variation in sepal lengths, with points spread out even more widely.

```python
#barplot
sns.barplot(x='species',y='sepal_length',data=iris)
plt.show()
```

The bar plot shows the total sepal length for each species.
You can see which species has the highest total sepal length.
The height of each bar indicates the sum of sepal lengths for that species.
This helps in understanding the overall contribution of each species to the total sepal length.




```python
#countplot
sns.countplot(x='species',data=iris)
plt.show()
```

The height of the bar indicates the number of observations in that category.

In the Iris dataset, each species (Setosa, Versicolor, Virginica) has 50 observations. Therefore, the bars for all three species will be of equal height, indicating that the dataset is balanced in terms of species distribution.



```python
#heatmap
numerical_df=df.select_dtypes(include=['float64','int64'])
sns.heatmap(numerical_df.corr(),annot=True)

plt.show()
```

 The heatmap shows the correlation between different numerical variables in the Iris dataset.

  The intensity of the color indicates the strength of the correlation. For example, a strong positive correlation is shown in warm color, while a strong negative correlation is shown in a cooler color.



```python
#scatterplot
sns.scatterplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
plt.show()
```

Scatter plots can show the correlation between the two variables.

If the points tend to rise together (from lower left to upper right), it indicates a positive correlation.

If the points tend to fall together (from upper left to lower right), it indicates a negative correlation.

If the points are scattered without any clear pattern, it suggests no correlation.

1. **Setosa**: Typically forms a distinct cluster, easily separable from the other two species.
Correlation: There is a slight negative correlation between sepal length and sepal width for Setosa, meaning as the sepal length increases, the sepal width slightly decreases.

2. **Versicolor**:Versicolor flowers are more spread out compared to Setosa, with moderate sepal lengths and widths.

Correlation: There is a weak positive correlation between sepal length and sepal width for Versicolor, indicating a slight increase in sepal width with increasing sepal length.

3. **Virginica**:Virginica flowers also form a distinct cluster but with longer sepal lengths and narrower sepal widths compared to Setosa.

Correlation: There is a weak positive correlation between sepal length and sepal width for Virginica, similar to Versicolor.

O**verall Correlation**
Sepal Length vs. Sepal Width: The overall correlation between sepal length and sepal width across all species is weak. This means that knowing the sepal length does not strongly predict the sepal width.

Species Differentiation: The scatter plot helps in differentiating between the species based on their sepal dimensions. Setosa is easily separable from Versicolor and Virginica, while Versicolor and Virginica overlap more.

```python
#lineplot
sns.lineplot(x='sepal_length',y='sepal_width',hue='species',data=iris)
plt.show()
```

The line plot shows trend in sepal length and sepal width across different species
