````python
d=pd.read_csv('/content/diabetes_model.csv')
d_df=pd.DataFrame(d)
d_df.head()
````

##Univariate Analysis:

````python
#1
plt.figure(figsize=(8,6))
plt.hist(d_df["BMI"],bins=5)
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('Distribution of BMI')
````

Histogram of BMI Column provides a visual representation of distribution of BMI in the dataset. Analysing the graph we can see that maximum distribution lies in the BMI range between 30-40.

````python
#2
plt.figure(figsize=(8,6))
plt.hist(d_df["Age"],bins=10)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
````

The above histogram represents the distribution of different age categories in the dataset. We can see that most data comes from age group of22-25(apprx)

##Bivariate Analysis

````python
#1
sns.scatterplot(x='Age',y='BMI',hue='Outcome',data=d_df)
plt.title('Age vs BMI')
plt.show()
````

As age increases, BMI tends to increase as well. This suggests a positive correlation between age and BMI.

````python
#2


sns.scatterplot(x='Age',y='BloodPressure',hue='Outcome',data=d_df)

plt.xlabel('Age')
plt.ylabel('Blood pressure')
plt.title('Age vs Blood pressure')
plt.show()
````

The scatterplot suggests a weak positive association between age and systolic blood pressure. Specifically:

As age increases, systolic blood pressure tends to rise slightly.

Keep in mind that other factors may also influence blood pressure, so this scatterplot provides a snapshot of one aspect of the relationship.

````python
#3

sns.scatterplot(x='Glucose',y='DiabetesPedigreeFunction',hue='Outcome',data=d_df)
plt.xlabel('Glucose')
plt.ylabel('DiabetesPedigreeFunction')
plt.title('DiabetesPedigreeFunction vs Glucose')
````

As Glucose levels increase, the DiabetesPedigreeFunction tends to be higher. This suggests that individuals with higher glucose levels may have a stronger genetic predisposition to diabetes.

````python
#4
corr=d_df.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
````

The above graph is a heatmap showing correlation between different variables. Values close to 1 shows +ve correlation and close to -1 shows -ve correlation.
We used the 'coolwarm' color scheme to map correlation.Color scheme close to warm tone shows +ve correlation and cool tone shows -ve correlation.
